# P-value, insight

This notebook is just a small visualisation of the p-value and the sensitivity results

We define a CDF of the $\Delta \Chi^{2}$ and its degrees of freedom to determine the p-value by:

$$p =  1 - CDF(\Delta \chi^{2}, dof_{\Delta \chi^{2}})$$

to find the sensitivity of the cosmologies.

```python
import numpy as np
import pylab as plt

# Plot a cummulative distribution function (CDF) to visually indicate the p-value test
x = np.linspace(0, 1, 100)
dist_func = 1 - (1/(1 + np.exp(-10 * (x - 0.5))))
invertes_dist_func = 1/(1 + np.exp(-10 * (x - 0.5)))


# alphas = list(zip(["red" if x >= 0 else 'red'], 0.25*dist_func))
# alphas_rising = list(zip(["green" if x >= 0 else 'green'], 0.25*invertes_dist_func))


plt.figure()
plt.title('P-value test', weight='bold', fontsize=16)
plt.plot(x, dist_func, color='k')
plt.fill_betweenx(dist_func, 0, 0.6, color='red', alpha=0.25, label='not-distinguishable', step='pre')
plt.fill_betweenx(dist_func, 0.8, 1, color='green', alpha=0.25, label='Distinguishable', step='post')
plt.fill_between(x, dist_func, 1, color='white')
# plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel(r'$P-value$', fontsize=14)
plt.xticks([])
plt.legend(loc='upper right')
plt.show()
```
